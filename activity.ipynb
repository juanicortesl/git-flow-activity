```python
import pandas as pd
import json
raw_tweets = pd.read_json(r'farmers-protest-tweets-2021-03-5.json', lines=True, nrows=100000)
print("Shape: ", raw_tweets.shape)
raw_tweets.head(5)
```

```python
def getTopTenRetweeted():
    topTen = raw_tweets.sort_values(by="retweetCount", ascending=False)
    topTen = topTen.iloc[:10]
    return topTen
 

topTen = getTopTenRetweeted()
topTen.head(10)
```

```python
def getUserName(row):
    return row["user"]['username']
def getTopTenUsers():
    topTen = raw_tweets
    topTen['userName'] = topTen.apply(lambda row:getUserName(row), axis=1)
    topTen = topTen.groupby(['userName']).count()
    topTen = topTen.sort_values(by="retweetCount", ascending=False)
    return topTen
topTen = getTopTenUsers()
topTen.head(10)
```

```python
import datetime
def getTopTenDays():
    topTen = raw_tweets
    topTen['day'] = pd.to_datetime(topTen['date']).dt.date
    topTen = topTen.groupby(['day']).count()
    topTen = topTen.sort_values(by="retweetCount", ascending=False)
    return topTen
topTen = getTopTenDays()
topTen.head(10)
```
